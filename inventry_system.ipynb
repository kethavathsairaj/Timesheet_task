```python
product_ids = []


categories = set()


inventory = {}

def add_product():
    pid = input("Enter Product ID: ")

    if pid in inventory:
        print("Product ID already exists")
        return

    name = input("Enter Product Name: ")
    category = input("Enter Category: ")
    quantity = int(input("Enter Quantity: "))
    price = float(input("Enter Price: "))

    product_info = (name, category)


    product_ids.append(pid)
    categories.add(category)


    inventory[pid] = {
        "info": product_info,
        "quantity": quantity,
        "price": price
    }

    print("Product Added Successfully")



def view_inventory():
    if not inventory:
        print("Inventory is empty")
        return

    for pid in inventory:
        name, category = inventory[pid]["info"]
        quantity = inventory[pid]["quantity"]
        price = inventory[pid]["price"]

        print("\nProduct ID:", pid)
        print("Name:", name)
        print("Category:", category)
        print("Quantity:", quantity)
        print("Price:", price)



def search_product():
    pid = input("Enter Product ID: ")

    if pid in inventory:
        name, category = inventory[pid]["info"]
        quantity = inventory[pid]["quantity"]
        price = inventory[pid]["price"]

        print("\nProduct Found")
        print("Name:", name)
        print("Category:", category)
        print("Quantity:", quantity)
        print("Price:", price)
    else:
        print("Product not found")

def show_categories():
    print("Categories:", categories)



while True:

    print("\n--- Inventory System ---")
    print("1 Add Product")
    print("2 View Inventory")
    print("3 Search Product")
    print("4 Show Categories")
    print("5 Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_product()

    elif choice == "2":
        view_inventory()

    elif choice == "3":
        search_product()

    elif choice == "4":
        show_categories()

    elif choice == "5":
        print("Exiting Program")
        break

    else:
        print("Invalid choice")
```
