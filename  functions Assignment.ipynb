```python
numbers=[2,-1,5,2,3,-7]
def positive_numbers(numbers):
    total=0
    for num in numbers:
        if num>0:
            total+=num
    return total
positive_numbers(numbers)
```

```python
x="madam"
def palindrome(x):
    return x==x[::-1]
palindrome(x)
```

```python
x=["banana","kiwi","apple","mango"]
def string(x):
    return sorted(x)
string(x)
```

```python
x=[1,2,3,4,5,6]
def square(x):
    squared_list = []
    for i in x:
        squared_list.append(i ** 2)
    return squared_list
print(square(x))
```
